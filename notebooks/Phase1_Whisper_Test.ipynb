```python
# Step 2  : Import Whisper
import whisper

# Step 3 : Load Model
model = whisper.load_model("base")

# Step 4 : Provide Audio
audio_path = "../uploads/sample.mp3"

# Step 5 : Transcribe
result = model.transcribe(audio_path)

# Step 6 : Display Transcript
print(result["text"])

# Step 7 : Save Transcript

with open(
    "../outputs/transcript.txt",
    "w",
    encoding="utf-8"
) as f:

    f.write(result["text"])
```
